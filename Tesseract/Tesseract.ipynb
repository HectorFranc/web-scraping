## Tesseract. OCR (Optical character recognition)
---

```python
import tesserocr # Para Optical character recognition
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
texto_largo = plt.imread('texto_largo.png')
plt.figure(figsize=(9, 4))
plt.imshow(texto_largo)
plt.axis(False)
plt.show()
```

## Texto desde archivo
---

```python
texto_ocr = tesserocr.file_to_text('texto_largo.png', lang='spa') # Converitr img a texto
```

```python
print(texto_ocr)
```

## Imagen no reconocida
---

```python
img = plt.imread('imagen de prueba.png')
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
texto_ocr = tesserocr.file_to_text('imagen de prueba.png', lang='spa')
print(texto_ocr) # No correct
```

**Sin canal alpha**

```python
img.shape # Alpha channel
```

```python
img_rgb = img[:, :, :3]
img_rgb.shape
```

```python
plt.imshow(img_rgb)
plt.show()
```

**Invertida (texto negro sobre fondo blanco)**

```python
img_inv = 1 - img_rgb
plt.imshow(img_inv)
plt.show()
```

**Blanco y negro**

```python
img_gray = img_inv.mean(axis=2)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

**Convertir en imagen de Pillow**

```python
img_pil = Image.fromarray(np.uint8(img_gray * 255))
```

```python
texto_ocr = tesserocr.image_to_text(img_pil, lang='spa')
print(texto_ocr) # Un error
```

```python
img_pil_inv = Image.fromarray(np.uint8(img_inv * 255))
texto_ocr = tesserocr.image_to_text(img_pil_inv, lang='spa')
print(texto_ocr) # Con imagen sin hacerla a escala de grises no hubo errores!!
```
